```python
rows_with_zero_xyz = diamonds.loc[(diamonds['x'] == 0) | (diamonds['y'] == 0) | (diamonds['z'] == 0)]
rows_with_zero_xyz
```

```python
carat_ix, table_ix, prince_ix, x_ix, y_ix, z_ix =1 ,6 ,7 ,8 ,9 ,10

from sklearn.base import BaseEstimator, TransformerMixin

class DiamondFeatureTransformer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X, y=None):
        X_transformed = X.copy()

        price_per_carat = X_transformed[:, prince_ix] / X_transformed[:, carat_ix]
        volume = (X_transformed[:,x_ix] * X_transformed[:,y_ix] * X_transformed[:,z_ix]) * (1/3)
        return np.c_[X, price_per_carat, volume]

attr_adder = DiamondFeatureTransformer()
diamond_extra_attribs = attr_adder.transform(diamonds.values)
diamond_extra_attribs
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

diamonds_num = diamonds.drop(['color', 'clarity', 'cut'], axis=1).reset_index(drop = True)

num_pipeline = Pipeline([
    ('attribs_adder', DiamondFeatureTransformer()),
    ('std_scaler', StandardScaler()),
])

diamonds_num_tr = num_pipeline.fit_transform(diamonds_num)
```

```python
from sklearn.compose import ColumnTransformer
num_attribs = list(diamonds_num)
cut_clarity_attribs = diamonds[['cut','clarity']]
color_attrib = diamonds[['color']]
one_hot = OneHotEncoder()
ordinal_encoder =  OrdinalEncoder()
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
    ("color_cat", OneHotEncoder(), color_attrib),
    ("cut_clarity_cat", OrdinalEncoder(), cut_clarity_attribs),
])
housing_prepared = full_pipeline.fit_transform(diamonds)
```
